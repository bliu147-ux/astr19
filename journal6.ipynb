```python
import numpy as np
from tabulate import tabulate
def cos(x):
    return np.cos(x)
def sin(x):
    return np.sin(x)
```

cos(x) function returns cos(x) from numpy, same with sin(x) but it instead returns sin(x)

```python
x = np.linspace(0,2*np.pi,1000)
y = cos(x)
z = sin(x)
table_data = [(a,b,c) for a,b,c in zip(x,y,z)]
table_headers = ["x","cos(x)", "sin(x)"]
python_table = tabulate(table_data, tablefmt="grid", headers=table_headers, floatfmt=".3f")
print(python_table)
```

```python
print("x", "cos(x)", "sin(x)")
for i in range(10):
    print(i, f'{cos(i):.3f}', f'{sin(i):.3f}')
```
