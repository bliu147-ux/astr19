```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,100)
def sine(x):
    return np.sin(x)                      #define the functions
def cosine(x):
    return np.cos(x)
y1 = sine(x)                       #assign them to variables for plotting
y2 = cosine(x)
```

```python
fig, axes = plt.subplots(1,2)              #create the subplots with fig and axes as the names to call from
axes[0].plot(x,y1,label='sin(x)')          #first plot to be labelled as sin(x)
axes[1].plot(x,y2,label='cos(x)')           #second to be labelled as cos(x)
axes[0].set_xlabel('x')               #individually setting the x-axis label
axes[1].set_xlabel('x')
axes[0].legend(loc=0)                 #setting the location of the legend, which uses information given by label='sin(x
axes[1].legend(loc=0)
plt.savefig('journal8.pdf',bbox_inches='tight',dpi=600)
plt.show()
```
